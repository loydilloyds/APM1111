```python
import statistics
import scipy
import math
import numpy

# Defining functions for std error of skewness and kurtosis
def se_skew(x):
    data_len = len(x)
    numerator = 6 * data_len * (data_len - 1)
    denom = (data_len - 2) * (data_len + 1) * (data_len + 3)
    stdskew = math.sqrt((numerator / denom))
    return stdskew

def se_kurt(x):
    data_len = len(x)
    stdskew = 2 * se_skew(x)
    numerator = (data_len*data_len) - 1
    denom = (data_len - 3) * (data_len + 5)
    stdkurt = stdskew * math.sqrt((numerator / denom))
    return stdkurt

val_set =   [
                88, 45, 53, 86, 33, 86, 85, 30, 89, 53, 41, 96, 56, 38, 62,
                71, 51, 86, 68, 29, 28, 47, 33, 37, 25, 36, 33, 94, 73, 46,
                42, 34, 79, 72, 88, 99, 82, 62, 57, 42, 28, 55, 67, 62, 60,
                96, 61, 57, 75, 93, 34, 75, 53, 32, 28, 73, 51, 69, 91, 35,
            ]

valid = len(val_set)
arr_mean = round(statistics.mean(val_set), 3)
arr_median = round(statistics.median(val_set), 3)
arr_mode = min(statistics.multimode(val_set))
arr_variance = round(statistics.variance(val_set), 3)
arr_stddev = round(statistics.stdev(val_set), 3)
arr_skew = round(scipy.stats.skew(val_set, axis = 0, bias = False), 3)
arr_kurt = round(scipy.stats.kurtosis(val_set, axis = 0, fisher=True, bias=False), 3)
arr_stdskew = round(se_skew(val_set), 3)
arr_stdkurt = round(se_kurt(val_set), 3)
arr_max = round(float(max(val_set)), 3)
arr_min = round(float(min(val_set)), 3)
arr_q1 = numpy.quantile(val_set, 0.25)
arr_q2 = numpy.quantile(val_set, 0.50)
arr_q3 = numpy.quantile(val_set, 0.75)
arr_d9 = max(numpy.percentile(val_set, numpy.arange(0, 100, 10)))
arr_p95 = max(numpy.percentile(val_set, numpy.arange(0, 100, 95)))

print(valid)
print(arr_mean)
print(arr_median)
print(arr_mode)
print(arr_variance)
print(arr_stddev)
print(arr_skew)
print(arr_stdskew)
print(arr_kurt)
print(arr_stdkurt)
print(arr_max)  
print(arr_min)
print(arr_q1)
print(arr_q2)
print(arr_q3)
print(arr_d9)
print(arr_p95)
```

```python
# Creating the table

import pandas as pd

tbl_index = ['Valid', 'Mode', 'Median', 'Mean', 'Std. Deviation', 'Variance', 'Skewness', 'Std. Error of Skewness', 'Kurtosis', 'Std. Error of Kurtosis', 'Minimum', 'Maximum', '25th percentile', '50th percentile', '75th percentile', '90th percentile', '95th percentile']
tbl_value = [valid, arr_mode, arr_median, arr_mean, arr_stddev, arr_variance, arr_skew, arr_stdskew, arr_kurt, arr_stdkurt, arr_min, arr_max, arr_q1, arr_q2, arr_q3, arr_d9, arr_p95]

data = pd.DataFrame(tbl_value, index=tbl_index, columns=['Scores'])
styler = data.style

border_props = [{'selector': 'tbody',
                 'props': [('border', '2px solid grey')]}]
styler.set_table_styles(border_props)
styler.format(precision=3)

display(styler)
```
